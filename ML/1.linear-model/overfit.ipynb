```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

np.random.seed(666)
x = np.random.uniform(-3,3,100)
X = x.reshape(-1, 1)

y = 0.5* x**2 + x+2 +np.random.normal(0,1,size = 100)

# create model
estimator = LinearRegression()
# train model
estimator.fit(X, y)
# use model to predict
y_predict = estimator.predict(X)

plt.scatter(x, y) # draw origin data
plt.plot(x, y_predict, color='r') # draw the line from model
plt.show()

# calculate MSE
from sklearn.metrics import mean_squared_error
mean_squared_error(y, y_predict)
```

X2中的每个样本都有两个特征。这样做的目的是为了引入更多的非线性关系来改善模型的表达能力。

```python
X2 = np.hstack([X, X**2])

estimator2 = LinearRegression()
estimator2.fit(X2, y)
y_predict2 = estimator2.predict(X2)

plt.scatter(x, y) # draw origin data
plt.plot(np.sort(x), y_predict2[np.argsort(x)], color = 'r') # draw line that fitted by model 
plt.show()

mean_squared_error(y, y_predict2)
```

再次加入高次项，绘制图像，观察均方误差结果

```python
X5 = np.hstack([X2,X**3,X**4,X**5,X**6,X**7,X**8,X**9,X**10])

estimator3 = LinearRegression()
estimator3.fit(X5, y)
y_predict5 = estimator3.predict(X5)

plt.scatter(x, y)
plt.plot(np.sort(x), y_predict5[np.argsort(x)], color='r')
plt.show()

error = mean_squared_error(y, y_predict5)
error
```
